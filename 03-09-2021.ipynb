### Matplotlib
- It is a crossplatform data visulization library for making 2d plots for given data.
- It is a one of the datascience package
- Types of plots:
    - Line plot
    - Bar graph
    - Scatterplot
    - Area plot
    - Pie Chart

```python
import matplotlib.pyplot as plt
```

```python
#Line plot
plt.plot([1,2,3,4],[2,4,6,8])
plt.title('xy-graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
#Bargraph- used to create bar chart
marks=[88,94,99,80,72]
names=['c','c++','python','java','oracle']
plt.bar(names,marks,color=['red','yellow','pink','green','black'],width=0.5)
```

```python
#Scatter plot
plt.scatter([1,2,3,4],[2,4,6,8],color='black')
plt.show()
```

```python
#Area plot
days=[1,2,3,4]
working=[8,11,9,10]
sleeping=[7,6,8,9]
plt.stackplot(days,working,sleeping,colors=['red','green'],labels=['work','rest'])
plt.legend()
plt.show()
```

```python
#pie chart
rating=[5,4,4.8,2]
names=['python','java','c','c++']
plt.pie(rating,labels=names)
plt.legend()
plt.show()
```
